```
N=20
a=-2
b=-2
h=(a+b)/N

fp=zeros([N,N],float)
xlim=linspace(-5,5,N)
xlim=linspace(-7,7,N)

for i in range (-N,N,1):
  x=xlim[i]
  for k in range (-N,N,1):
    y=xlim[k]
    R=0
    for j in range (1,N,2):
      R=4*f((a+j*h),y)

    for j in range (2,N,2):
      R=2*f((a+j*h),y)
    R=f(a,y)+f(b,y)+(1/3)*h*R
    E=8.99e9*5*sqrt(R**2)
    fp[j,i]=E
```
